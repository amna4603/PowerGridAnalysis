```python

X_G1_pu = 0.12
X_G2_pu = 0.12
X_T1_pu = 0.1
X_T2_pu = 0.1
Z_L23_pu =  1j*0.0142
Z_L34_pu =  1j*0.035916824
Z_L45_pu =  1j*0.037807
Z_L46_pu =  1j*0.014177694
Z_L56_pu =  1j*0.047258979

#starToDeltaTransformation
Z4g_pu = 1j*X_G1_pu + 1j*X_T1_pu + Z_L23_pu + Z_L34_pu  
Z6g_pu = 1j*(X_G2_pu + X_T2_pu)
Z45_delta_pu = (Z_L46_pu*Z_L56_pu + Z_L46_pu*Z6g_pu + Z_L56_pu*Z6g_pu)/Z6g_pu
Z5g_delta_pu = (Z_L46_pu*Z_L56_pu + Z_L46_pu*Z6g_pu + Z_L56_pu*Z6g_pu)/Z_L46_pu
Z4g_delta_pu = (Z_L46_pu*Z_L56_pu + Z_L46_pu*Z6g_pu + Z_L56_pu*Z6g_pu)/Z_L56_pu

Z45_eqv_pu = 1/(1/Z45_delta_pu + 1/Z_L45_pu)
Z4g_eqv_pu = 1/(1/Z4g_delta_pu + 1/Z4g_pu)
Z5g_eqv_pu = 1/(1/(Z45_eqv_pu + Z4g_eqv_pu) + 1/Z5g_delta_pu)

Zth = Z5g_eqv_pu

print("Thevenin equivalent impedance at bus 5 is: ", Zth)


```

```python

#Positive sequence
X_G1_1 = 1j*0.12
X_G2_1 = 1j*0.06
X_T1_1 = 1j*0.1
X_T2_1 = 1j*0.05
Z_L23_1 =  1j*0.014177694
Z_L34_1 =  1j*0.035916824
Z_L45_1 =  1j*0.037807
Z_L46_1 =  1j*0.014177694
Z_L56_1 =  1j*0.047258979

#Negative sequence
X_G1_2 = 1j*0.14
X_G2_2 = 1j*0.07
X_T1_2 = 1j*0.1
X_T2_2 = 1j*0.05
Z_L23_2 =  1j*0.014177694
Z_L34_2 =  1j*0.035916824
Z_L45_2 =  1j*0.037807
Z_L46_2 =  1j*0.014177694
Z_L56_2 =  1j*0.047258979

#Zero sequence
X_G1_0 = 1j*0.05
X_G2_0 = 1j*0.025
X_T1_0 = 1j*0.1
X_T2_0 = 1j*0.05
Z_L23_0 =  1j*0.042533081
Z_L34_0 =  1j*0.107750473
Z_L45_0 =  1j*0.11342155
Z_L46_0 =  1j*0.042533081
Z_L56_0 =  1j*0.141776938

def calculate_parallel(Z1,Z2):
    return Z1*Z2/(Z1+Z2)

def star2delta(Z1, Z2, Z3):

    numerator = Z1*Z2 + Z1*Z3 + Z2*Z3

    Za = numerator/Z2
    Zb = numerator/Z3
    Zc = numerator/Z1
    
    return Za, Zb, Zc

#Positive thevenin equivalent for fault at bus 5

Z_bus4_1 = X_G1_1+X_T1_1+Z_L23_1+Z_L34_1
Z_bus6_1 = X_G2_1+X_T2_1

Za,Zb,Zc = star2delta(Z_L46_1,Z_bus6_1,Z_L56_1)

Z1 = calculate_parallel(calculate_parallel(Z_bus4_1,Zb)+calculate_parallel(Za,Z_L45_1),Zc)

print("Thevenin equivalent impedance at bus 5 is: ", Z1)

#Negative thevenin equivalent for fault at bus 5

Z_bus4_2 = X_G1_2+X_T1_2+Z_L23_2+Z_L34_2
Z_bus6_2 = X_G2_2+X_T2_2

Za,Zb,Zc = star2delta(Z_L46_2,Z_bus6_2,Z_L56_2)

Z2 = calculate_parallel(calculate_parallel(Z_bus4_2,Zb)+calculate_parallel(Za,Z_L45_2),Zc)

print("Thevenin equivalent impedance at bus 5 is: ", Z2)

#Zero thevenin equivalent for fault at bus 5

Z_bus4_0 = X_T1_0+Z_L23_0+Z_L34_0
Z_bus6_0 = X_T2_0

Za,Zb,Zc = star2delta(Z_L46_0,Z_bus6_0,Z_L56_0)

Z0 = calculate_parallel(calculate_parallel(Z_bus4_0,Zb)+calculate_parallel(Za,Z_L45_0),Zc)

print("Thevenin equivalent impedance at bus 5 is: ", Z0)
```
